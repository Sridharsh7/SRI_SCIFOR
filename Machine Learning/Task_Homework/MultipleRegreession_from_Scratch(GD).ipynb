### **Multiple linear regression** is a method to model the relationship between one dependent variable and two or more independent variables.

---



### Code from Scratch to Implement Multiple Linear Regression:

---



```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Preparing the dataset
# Example dataset
# Features: X1, X2
# Target: y
X = np.array([
    [1, 2],
    [2, 3],
    [4, 2],
    [3, 4],
    [5, 5]
])

y = np.array([5, 7, 10, 11, 14])

# Adding a column of ones to X for the intercept term (bias)
X = np.c_[np.ones(X.shape[0]), X]

# 2. Implementing the multiple linear regression model
def predict(X, theta):
    return np.dot(X, theta)

# 3. Training the model using gradient descent
def gradient_descent(X, y, learning_rate=0.01, iterations=1000):
    m, n = X.shape
    theta = np.zeros(n)
    for _ in range(iterations):
        gradient = (1/m) * np.dot(X.T, (predict(X, theta) - y))
        theta -= learning_rate * gradient
    return theta

# Hyperparameters
learning_rate = 0.01
iterations = 1000

# Train the model
theta = gradient_descent(X, y, learning_rate, iterations)

# 4. Evaluate the model
y_pred = predict(X, theta)

# Calculate Mean Squared Error (MSE)
mse = np.mean((y_pred - y) ** 2)
# Calculate Mean Absolute Error (MAE)
mae = np.mean(np.abs(y_pred - y))
# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate R-squared (R²)
ss_total = np.sum((y - np.mean(y)) ** 2)
ss_residual = np.sum((y - y_pred) ** 2)
r2 = 1 - (ss_residual / ss_total)

print("Theta (coefficients):", theta)
print("Predicted values:", y_pred)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
print("R-squared (R²):", r2)
```

```python
# 6. Plotting
plt.figure(figsize=(10, 6))
plt.scatter(range(len(y)), y, color='blue', label='Actual values')
plt.plot(range(len(y)), y_pred, color='red', linestyle='--', label='Predicted values')
plt.xlabel('Sample index')
plt.ylabel('Target value')
plt.title('Comparison of Actual and Predicted Values')
plt.legend()
plt.show()
```
