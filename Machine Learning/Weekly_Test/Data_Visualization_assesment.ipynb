1. Choose a dataset and load a dataset

```python
import pandas as pd

# Load the dataset
url = '/content/titanic.csv'
titanic = pd.read_csv(url)

# Inspect the dataset
print(titanic.head())
print(titanic.info())
```

2. Handling Missing Values

```python
# Remove rows with missing values
titanic_cleaned = titanic.dropna()

# Verify that there are no missing values
print(titanic_cleaned.info())
```

3. Visualizations using Bar chart

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart: Number of passengers per class
plt.figure(figsize=(10, 5))
sns.countplot(x='Pclass', data=titanic_cleaned)
plt.title('Number of Passengers per Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

4. Visualizations using Pie chart

```python
# Pie chart: Survival rate
survival_counts = titanic_cleaned['Survived'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(survival_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=140)
plt.title('Survival Rate')
plt.show()
```

5. Visualization Using Line chart

```python
# Line chart: Age distribution by class
plt.figure(figsize=(10, 5))
sns.lineplot(x='Pclass', y='Age', data=titanic_cleaned)
plt.title('Age Distribution by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

6. Data Visualization Using Histogram

```python
# Histogram: Age distribution
# Histogram: Age distribution
plt.figure(figsize=(10, 5))
sns.histplot(titanic_cleaned['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
